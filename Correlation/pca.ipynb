```python
import pandas as pd

df = pd.read_csv('../Datasets/states/Kerala.csv')
df_agg = df.groupby(['Date', 'District']).agg({'Avg_rainfall': 'sum'}).reset_index()
pivot_df = df_agg.pivot(index='Date', columns = 'District', values='Avg_rainfall')
pivot_df.reset_index(inplace=True)
pivot_df.columns.name = None
pivot_df.set_index('Date', inplace=True)
pivot_df.reset_index(drop=True, inplace=True)
pivot_df = pivot_df.fillna(0)
total_df = pivot_df
pivot_df
```

Finding Pearson Correlations Matrix

```python
corr = pivot_df.corr(method='pearson')
corr
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(15,6), dpi =500)
sns.heatmap(corr,annot=True,fmt=".2f", linewidth=.5)
plt.show()
```

Principal Component Analysis: 
Standardize the data.
Use the standardized data to generate a covariance matrix (or perform Singular Vector Decomposition).
Obtain eigenvectors (principal components) and eigenvalues from the covariance matrix. Each eigenvector will have a corresponding eigenvalue.
Sort the eigenvalues in descending order.
Select the k eigenvectors with the largest eigenvalues, where k is the number of dimensions used in the new feature space (kâ‰¤d).
Construct a new matrix with the selected k eigenvectors.

```python
import pandas as pd
import numpy as np

pivot_df.head()
```

```python
def data_normalisation(scaled_data):
    for col in scaled_data.columns: 
        scaled_data[col] = (scaled_data[col]-scaled_data[col].mean())/scaled_data[col].std() #data normalisation 
    
    return scaled_data

scaled_data = data_normalisation(pivot_df)
print(scaled_data)
```

```python
def covariance_calculation(mean_subtracted):
    #calcualte covariance amongst scaled values
    for col in mean_subtracted.columns:
        mean_subtracted[col] = mean_subtracted[col]-mean_subtracted[col].mean()  
    return np.dot(mean_subtracted.T,mean_subtracted)/(len(mean_subtracted) - 1)
cov_calc = covariance_calculation(scaled_data) 
```

```python
print('Shape of scaled data:',scaled_data.shape)
print('Shape of covariance matrix:',cov_calc.shape)
```

```python
import matplotlib.pyplot as plt
from numpy.linalg import eig
eig_values,eig_vectors=eig(cov_calc)

plt.figure(figsize=(5,3))
plt.stem(eig_values[0:])
plt.title('Eigen Values')
plt.xlabel('Number of Components')
plt.ylabel('Eigen Values')
plt.show()

# plt.bar(["e" + str(i+1) for i in range(len(eig_values))], eig_values)
# plt.title("Eigenvalues")
# plt.xlabel("Magnitude of the eigenvalue")
```

```python
explained_variance = [(i/sum(eig_values))*100 for i in eig_values]
explained_variance = np.round(explained_variance,2)
print('Explained variance: {}'.format(explained_variance))
for i in range(len(explained_variance)):
    exp_var = np.sum(eig_values[:i+1])*100 / np.sum(eig_values)
    print(f'Eigenvectors upto {i+1} expresses {exp_var} % variance')
```

```python
cumulative = np.cumsum(explained_variance)
plt.plot(np.arange(1,scaled_data.shape[1]+1), cumulative, marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance')
plt.xticks(np.arange(1,scaled_data.shape[1]+1))
plt.show()
```

```python
eig_pairs = [(np.abs(eig_values[i]), eig_vectors[:,i]) for i in range(len(eig_values))]
eig_pairs.sort(key=lambda x: x[0], reverse=True)
values = np.array(x[0] for x in eig_pairs)
vectors = np.array(x[1] for x in eig_pairs)
# print(eig_pairs)
```

```python
k = 2
proj_matrix = eig_vectors[:k,:]
print('Projection matrix shape:',proj_matrix.shape)
```

```python

projected1 = scaled_data.dot(eig_vectors.T[0])
projected2 = scaled_data.dot(eig_vectors.T[1])

projected_data = pd.DataFrame(projected1, columns = ['PC1'])
projected_data['PC2'] = projected2

print('Shape of original data:',scaled_data.shape)
print('Shape of projected data:',projected_data.shape)

projected_data
```

```python
def rainfall(state):
    df = pd.read_csv('../Datasets/states/'+state+'.csv')
    df.drop(columns=['State','Year','Month', 'Agency_name'], inplace=True)
    df['Date'] = pd.to_datetime(df['Date'])
    df = df.groupby('District')['Avg_rainfall'].sum().reset_index()
    df.sort_values(by='Avg_rainfall', ascending=False, inplace=True, )
    df.reset_index(drop=True, inplace=True)
    return df
```

```python
#plotting the data
import pandas as pd
import matplotlib.pyplot as plt
def plot_data(state):
    plt.figure(figsize=(15,6))
    plt.bar(state['District'], state['Avg_rainfall'])
    plt.xlabel('Districts')
    plt.ylabel('Total Rainfall')
    plt.title('Total Rainfall in each District')
    plt.xticks(rotation=90)
    plt.show()

data = rainfall('Kerala')
plot_data(data)
```
